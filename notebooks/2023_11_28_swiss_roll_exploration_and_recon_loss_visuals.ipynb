```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll

# Generate Swiss Roll dataset
# The data should be randomized already
n_samples = 10000
noise = 0.5
X, color = make_swiss_roll(n_samples=n_samples, noise=noise)
X = X 



# # shuffle data
# idx = np.arange(n_samples)
# np.random.shuffle(idx)
# X = X[idx]
# color = color[idx]


# Plot the Swiss Roll
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
ax.set_title('Swiss Roll')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()

# reshape data to have 100 trials with each trial having 100 samples
x_all = np.array(X.reshape(100, 100, 3).tolist(), dtype=object)
u_all = np.array(color.reshape(100, 100, 1).tolist(), dtype=object)
print(x_all.shape)
print(u_all.shape)

```

```python
swiss_roll_pivae_rec_loss = np.array([15.581, 15.69, 15.516, 15.518]) # with different seeds
swiss_roll_vae_rec_loss = np.array([18.1, 18.1, 18.0, 18.0]) # with different seeds

# compare the pivae and vae reconstruction loss
#make side by side barplot with error bars equals to std of the loss across seeds
plt.figure(figsize=(8, 6))
plt.bar(x=1, height=swiss_roll_pivae_rec_loss.mean(), width=0.4, yerr=swiss_roll_pivae_rec_loss.std(), label='PIVAE')
plt.bar(x=1.6, height=swiss_roll_vae_rec_loss.mean(), width=0.4, yerr=swiss_roll_vae_rec_loss.std(), label='VAE')
plt.title('Reconstruction loss (Rat Hippo)', fontsize=20)
plt.ylabel('Reconstruction loss', fontsize=15)
plt.legend(fontsize=20)
plt.show()
```
